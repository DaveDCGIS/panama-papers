# Panama Papers Graph Analysis

The Panama Papers data is comprised of five csv data sets from the International Consortium of Investigative Journalists (ICIJ) that includes entities, intermediaries, officers, addresse, and all edges that connectes all data sets.

Data (source: ICIJ)
1. Entities.csv
2. Intermediaries.csv
3. Officers.csv
4. Addresses.csv
5. All_Edges.csv

The purpose of this analysis is to create a visual analysis using the "all edges" data set

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

In order to perform our analysis, we first access the "all_edges" data set from Microsoft Azure cloud storage

```python
a = pd.read_csv("https://panamadata.blob.core.windows.net/icij/all_edges.csv")
```

Print the top ten records of the "all_edges" data set

```python
a.head(10)
```

Create a list based on the node_1 and node_2 available in the "all_edges" data set

```python
a1 = (a.node_1,a.node_2)
```

Print a sample record from the newly created list

```python
print(a1[0][6],a1[1][6])
```

Using a simple python loop create sets of edges from list a1 for entirety of data set

```python
g = nx.Graph()
i = 0
j = 0
k = 1

for j in range(len(a1[0])):
    g.add_edge(a1[i][j], a1[k][j])
    j = j+1
```

```python
print(nx.info(g))
```

```python
nx.degree_centrality(g)
```

## For the purposes of a simpler analysis, a sample of 1000 records were used below

```python
g1 = nx.Graph()
a, b = 0,0
c = 1

for j in range(1000):
    g1.add_edge(a1[a][b], a1[c][b])
    b = b+1
```

```python
print(nx.info(g1))
```

```python
nx.betweenness_centrality(g1)
```

```python
nx.draw(g1)
plt.show()
```

```python
nx.draw(g1, cmap = plt.get_cmap('jet'))
plt.show()
```

```python
g2 = nx.petersen_graph(g1)
nx.draw(g2)
```

```python
g3 = nx.karate_club_graph()
nx.draw(g3)
```

```python
#pos = nx.spring_layout(g1,scale=2)
pos = nx.spring_layout(g1)
#nx.draw_networkx_edges(g1, pos, font_size= 7, edge_color='red', arrows=True)
nx.draw_networkx_nodes(g1,pos,node_color='b',alpha=1,node_size=12)
nx.draw_networkx_edges(g1,pos,alpha=0.5)
```
